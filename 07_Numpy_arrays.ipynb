# Why to use arrays instead of list

```python
a = [3, 4, 5, 6, 7]
print(type(a))

print(a * 3)

# Intialize a blank list
b = []
for i in a:
    b.append(i * 3)

print(f"{a} * 3 = {b}")
```

```python
b
```

```python
a1 = np.array(a)
a1
```

```python
a1 * 3
```

```python
5 + a1 * 4 
```

```python
a1 ** 2
```

```python
a1 ** 3
```

```python
a1 ** 4
```

```python
a1 ** (1/2)
```

```python
a1 ** (1/3)
```

```python
a1 ** (1/4)
```
