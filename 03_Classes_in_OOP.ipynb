# Person Class


```python
class Person:

    def __init__(self, name: str, age: int, gender: str, occupation: str):
        self.name = name
        self.age = age
        self.gender = gender
        self.occupation = occupation

    def intro(self):
        print(f"My name is {self.name}")
        print(f"I am {self.age} years old")
        print(f"I am {self.gender}")
        print(f"I work as {self.occupation}")
```

```python
p1 = Person(name = "Rahul", age=28, gender="Male", occupation="Accounts")
print(type(p1))
```

```python
p1.occupation
```

```python
p1.intro()
```

```python
print(p1)
```

# Show the string representation \_\_str\_\_

```python
class Person2:

    def __init__(self, name: str, age: int, gender: str, occupation: str):
        self.name = name
        self.age = age
        self.gender = gender
        self.occupation = occupation

    def __str__(self):
        return f"Person2(name = {self.name}, age = {self.age}, gender = {self.gender}, occupation = {self.occupation})"

    def talk(self):
        print(f"My name is {self.name}")
        print(f"My age is {self.age}")
        print(f"I am {self.gender}")
        print(f"I work as a {self.occupation}")
```

```python
p1 = Person2(name="Raman", age=28, gender="Male", occupation = "Accounts")
print(p1)
```

```python
p1.name
```

```python
p1.age
```

```python
p1.gender
```

```python
p1.occupation
```

```python
type(p1)
```

```python
p1.talk()
```

# Dataclasses

```python
from dataclasses import dataclass
from typing import Literal

@dataclass
class Student:
    roll_no: int
    name: str
    age: int
    gender: Literal["male", "female"]

    def get_info(self):
        print(f"Roll no : {self.roll_no}")
        print(f"Name : {self.name}")
        print(f"Age : {self.age}")
        print(f"Gender : {self.gender}")
```

```python
s1 = Student(roll_no=102, name="Rahul", age=28, gender="male")
print(s1)
```

```python
print(type(s1))
```

```python
s2 = Student(roll_no=103, name="Raman", age=30, gender="male")
print(s2)
```

```python
s2.name
```

```python
s2.roll_no
```

```python
s2.age
```

```python
s2.gender
```

```python
s2.get_info()
```
