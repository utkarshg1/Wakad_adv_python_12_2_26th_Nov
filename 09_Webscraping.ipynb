# Webscraping

```python
import requests
from bs4 import BeautifulSoup
```

# Get html content from webpage

```python
import requests

url1 = "https://en.wikipedia.org/wiki/World_population"
response = requests.get(
    url1,
    headers = {"User-Agent": "UtkarshBot/1.0 (utkarsh@example.com)"}
)
response
```

```python
# HTML content will be stored here
content = response.content
```

# Get the title from the webpage

```python
soup = BeautifulSoup(response.content)
title_tag = soup.find("title")
print(title_tag.text)
```

# Get the h1 tag from the page

```python
h1_tag = soup.find("h1")
h1_tag.text
```

# Get all the subtitles from wikipedia webpage

```python
subheading_tags = soup.find_all("div", class_ = "mw-heading")
subheading_tags
```

```python
sub_text = [tag.text for tag in subheading_tags]
sub_text
```

```python
p_tags = soup.find_all("p")
p_tags[0:5]
```

```python
p_tags[1].text
```

```python
p_text = [tag.text for tag in p_tags]
p_text[0:5]
```

# Save all the paragraph content

```python
with open("world.txt", "w", encoding="utf-8") as file:
    file.writelines(p_text)
```

# Load the text file

```python
with open("world.txt", "r") as file:
    content = file.read()
```

```python
print(content[0:100])
```

# Get the image urls

```python
img_tags = soup.find_all("a", class_ = "mw-file-description")
img_tags[0:5]
```

```python
img_tags[0]
```

```python
home_page_url = "https://en.wikipedia.org"
home_page_url + img_tags[0].get("href")
```

```python
img_urls = [home_page_url + tag.get("href") for tag in img_tags]
img_urls
```
