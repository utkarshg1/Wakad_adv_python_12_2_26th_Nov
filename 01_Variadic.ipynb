# Variadic function - This provoides flexibility in input to users
1. esc, m - convert cell to markdown
2. esc, y - convert cell to code

```python
def add(a: int|float, b: int|float) -> int|float:
    """
    This function adds two numbers a and b
    """
    c = a + b
    return c
```

```python
# Positional arguments
add(3, 4)
```

```python
# Keyword arguments
add(a=12, b=13)
```

```python
add(12.5, 13.2)
```

```python
a1 = add(a=12.5, b=13.2)
print(a1)
```

```python
add(2, 3, 4)
```

```python
add(2, 3, 4, 5)
```

# Positional arguments with flexible input are repsented as *args inside function
\*args will act like a tuple

```python
def function1(*args):
    print(f"Postional Arguments : {args}")
    print(f"Type : {type(args)}")
    for i in args:
        print(i)
```

```python
function1(1, "Example")
```

```python
function1(2, 3, 4, "Test", "Example", True)
```

```python
function1(11, 12, 13, "Test")
```

# You can unpack the list using *list

```python
a = [1, 2, 3, "Test", 12.5, False, True]
print(a, type(a))
```

```python
function1(*a)
```

```python
b = [1, 2, 3]
function1(*b)
```

# Write a variadic function to add multiple numbers

```python
def add_multiple(*args) -> int | float:
    # Intialize the sum to 0 
    s = 0
    # Add the elements inside args
    for i in args:
        s = s + i
    return s
```

```python
a1 = add_multiple(2, 3)
print(a1)
```

```python
a2 = add_multiple(2, 3, 4, 5, 6)
print(a2)
```

```python
list(range(1, 10))
```

```python
nums = list(range(1, 1001))
print(nums)
a3 = add_multiple(*nums)
print(a3)
```

# Cart values

```python
"="*10
```

```python
cart = [
    ("Laptop", 2, 45000),
    ("Mouse", 1, 500),
    ("Keyboard", 2, 1500),
    ("Screen Guard", 3, 800)
]
```

```python
def billing(*args):
    total_bill = 0
    for name, qty, price in args:
        line_total = price * qty
        total_bill = total_bill + line_total
        print(f"Name : {name}, Quantity : {qty}, Price : {price}, Line Total : {line_total}")
        print("="*80)
    return total_bill
```

```python
cart
```

```python
b1 = billing(*cart)
print(f"Total Bill : {b1} INR")
```

```python
b2 = billing(
    ("Mobile", 2, 25000),
    ("Cover", 4, 300),
    ("Screen guard", 2, 400)
)
print(f"Total Bill : {b2} INR")
```

# Keyword arguments
**kwargs - these act like a dictionary in python

```python
def function2(**kwargs):
    print(f"Keyword arguments : {kwargs}")
    print(f"Type : {type(kwargs)}")
    for key, value in kwargs.items():
        print(f"Key : {key}, Value : {value}")
```

```python
function2(name="Rahul", age=28)
```

```python
function2(name="Raman", age=32, email="r@test.com", city="Chennai")
```

```python
function2(a=1, b=3, c="Test")
```

# You can unpack a dictionary using **dict

```python
d = {
    "id": 101,
    "first_name" : "Sarthak",
    "last_name": "More",
    "email": "sarthak@test.com"
}
print(d, type(d))
```

```python
function2(**d)
```

# Functions with positional and keyword arguments
1. *args - Positional Arguments - Tuple
2. **kwargs - Keyword arguments - Dictionary
3. Always positional arguments are first then keyword arguments

```python
def function3(*args, **kwargs):
    print(f"Postional Arguments : {args}")
    print(f"Type Positinal Arguments : {type(args)}")
    for i in args:
        print(i)
    print("="*60)
    print(f"Keyword Arguments : {kwargs}")
    print(f"Type Keyword Arguments : {type(kwargs)}")
    for key, value in kwargs.items():
        print(f"Key : {key}, Value : {value}")
```

```python
function3(1, 2, 3, 4, "Example", "Test", False)
```

```python
function3(a="Rahul", b=23, d=False, id=123)
```

```python
function3(1, 2, 3, "Example", True, a=12, b=13, c="Test", d=False)
```

```python
function3(a=12, b=13, 34, 12, 13)
```

```python
a = [12, 13, 14, "Example", "Test", False]
b = {"id": 103, "name": "Rahul", "age": 28}
function3(*a, **b)
```
